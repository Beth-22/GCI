```python
users = np.unique(data[:,0])
movies = np.unique(data[:,1])
print("Unique Users:", users)
```

```python
# creating index maps
users_to_index = {user: index for index, user in enumerate(users)}
movies_to_index = {movie: index for index, movie in enumerate(movies)}
```

```python
ratings_matrix = np.zeros((len(users), len(movies)), dtype=int)
for user, movie, rating in data:
    user_index = users_to_index[user]
    movie_index = movies_to_index[movie]
    ratings_matrix[user_index, movie_index] = rating
```

```python
print("Users:", users)
print("Movies:", movies)
print("Ratings Matrix:\n", ratings_matrix)
```

```python
#Average rating per movie
average_ratings = np.mean(ratings_matrix, axis=0)
for movie, avg in zip(movies, average_ratings):
    print(f"{movie.decode('utf-8')}: {avg}") 
    
# Sort movies by average rating
sorted_indices = np.argsort(average_ratings)[::-1]  # Sort in descending order
print("\nMovies sorted by average rating:")
for idx in sorted_indices:
    print(f"{movies[idx].decode('utf-8')}: {average_ratings[idx]}")
```

```python
#average rating per user
average_user_ratings = np.mean(ratings_matrix, axis=1)
for user, avg in zip(users, average_user_ratings):
    print(f"{user.decode('utf-8')}: {avg}") 
#sorted by average rating
sorted_user_indices = np.argsort(average_user_ratings)[::-1]  # Sort in descending order
print("\nUsers sorted by average rating:")
for idx in sorted_user_indices:
    print(f"{users[idx].decode('utf-8')}: {average_user_ratings[idx]}")
    if average_user_ratings[idx] < 4:
        print(f"{users[idx].decode('utf-8')} is a harsh critic")
```

```python
#top 3 movies
print("\nTop 3 Movies:")
for idx in sorted_indices[:3]:
    print(f"{movies[idx].decode('utf-8')}: {average_ratings[idx]}")
    
```

```python
#number of good ratings (4 or 5) per movie
good_ratings_count = np.sum(ratings_matrix >= 4, axis=0)
print("\nNumber of good ratings (4 or 5) per movie:")
for movie, count in zip(movies, good_ratings_count):
    print(f"{movie.decode('utf-8')}: {count}")

#sorted by good ratings
sorted_by_good_ratings_indices = np.argsort(good_ratings_count)[::-1]
print("\nMovies sorted by number of good ratings:") 
for idx in sorted_by_good_ratings_indices:
    print(f"{movies[idx].decode('utf-8')}: {good_ratings_count[idx]} good ratings")
        
```
